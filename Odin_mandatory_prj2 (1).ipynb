**California Housing Dataset**

The aim of california housing dataset is to predict median house value based on the parameters like longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,median_house_value,ocean_proximity

**Column name and their description**

**longitude**- The term used to measure how far the house is aligned to west side, higher the value higher the west side

**latitude**-  The term used to measure how far the house is aligned to North side, higher the value higher the North side


**housing_median_age**- Meadian age of a house

**total_rooms**- Total number of rooms within a block

**total_bedrooms**- Total number of bed rooms within a block

**population**- Total number of Population within a block

**households**- Total number of household within a block

**median_income**- Median income for People within a block of houses (measured in tens of thousands of USD)

**median_house_value**-Median house value for households within a block (measured in USD)

**ocean_proximity**- Location of the house with respect to ocean/sea

**Feature    and                 Type of feature**

longitude  -                Continuous

latitude -                  Continuous

housing_median_age -        Continuous

total_rooms -               Discrete

total_bedrooms  -           Discrete

population  -               Continuous

households  -               Continuous

median_income -             Discrete

median_house_value  -       Discrete

ocean_proximity   -         Nominal

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# 10)
import numpy as np
conditions=[
    (df['total_bedrooms']<=10),
    (df['total_bedrooms']>10) & (df['total_bedrooms']<=1000),
    (df['total_bedrooms']>1000)
]
values=['small','medium','large']
df['total_bedroom_size']=np.select(conditions, values) #matches the condition and create the extra colummn
df
```
